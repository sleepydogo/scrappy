```python
import pandas as pd
import re
import os
```

# Concatenador 

```python
# Esta funcion toma una carpeta y extrae de todas las hojas de datos 
# existentes el numero de telefono.

# Directorio donde se encuentran los archivos
directorio = '/home/tom/Desktop/programa/data/Colombia 21-01'
campo_nro_telefonico = 'Telefono'

# Lista para almacenar los marcos de datos de cada archivo
datos = []

# Leer cada archivo en el directorio
for archivo in os.listdir(directorio):
    if archivo.endswith('.xlsx') or archivo.endswith('.xls'):
        ruta_archivo = os.path.join(directorio, archivo)
        # Leer el archivo y extraer la columna 'telefono'
        df = pd.read_excel(ruta_archivo)
        if campo_nro_telefonico in df.columns:
            datos.append(df[campo_nro_telefonico])

# Concatenar todos los datos en un solo DataFrame
datos_concatenados = pd.concat(datos, ignore_index=True)

# Guardar los datos concatenados en un nuevo archivo
datos_concatenados.to_excel('datos_concatenados.xlsx', index=False)

print("Datos concatenados y guardados en 'datos_concatenados.xlsx'")
```

# Comparador

```python
# Cargar los archivos Excel en DataFrames de Pandas
datos_unicos = pd.read_excel("datos_concatenados.xlsx")
prospectos = pd.read_excel("/home/tom/Desktop/programa/data/Colombia 27-03/taxi_Colombia_prospectos.xlsx")
```

```python

# Esta funcion transforma los numeros de telefono "+51 942 407 403" --> "51942407403"
def limpiar_telefono(numero):
    # Verificar si el valor es una cadena
    if isinstance(numero, str):
        # Utilizar una expresión regular para eliminar caracteres no deseados
        numero_limpio = re.sub(r'\D', '', numero)
        # Si el número tiene el prefijo internacional, quitarlo
        return numero_limpio
    else:
        # Si no es una cadena, devolver el mismo valor
        return numero
    
# Transformamos los telefonos en formato string
datos_unicos['Telefono'] = datos_unicos['Telefono'].astype(str)
prospectos['Telefono'] = prospectos['Telefono'].astype(str)

# Aplicamos el filtro de los numeros
datos_unicos['Telefono'] = datos_unicos['Telefono'].apply(limpiar_telefono)
prospectos['Telefono'] = prospectos['Telefono'].apply(limpiar_telefono)

print('Datos nuevos \n', datos_unicos['Telefono'])
print('Datos antiguos \n', prospectos['Telefono'])
```

```python
telefonos_nuevos = prospectos[~prospectos['Telefono'].isin(datos_unicos['Telefono'])]

# Guardar los resultados en un nuevo archivo
telefonos_nuevos.to_excel('telefonos_nuevos.xlsx', index=False)

print("Telefonos únicos guardados en 'telefonos_nuevos.xlsx'")
```
